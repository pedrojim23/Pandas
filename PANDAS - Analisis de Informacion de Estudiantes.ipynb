```python
#Como primer paso importamos las librerias necesarias para el ejercicio
import pandas as pd #Para manipulación y análisis de datos
import matplotlib.pyplot as plt #Para crear gráficos básicos
import seaborn as sns #Para visualizaciones estadísticas más sofisticadas
```

```python
#Cargamos el dataset de Informacion de Estudiantes
df = pd.read_csv('Cleaned_Students_Performance.csv')

#Mostramos las primeras filas del dataset para corroborar que se subio correctamente
df.head()
```

```python
#erifica la cantidad de valores nulos en cada columna
print(df.isnull().sum())

#Proporciona información detallada sobre el dataset
df.info()

#Genera estadísticas descriptivas para columnas numéricas
df.describe()
```

```python
#Establece un estilo de fondo de cuadrícula blanca
sns.set(style = "whitegrid")

#Configura el tamaño del gráfico a un tamaño de 10x6 y generamos un diagrama de cajas
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math_score', data=df, palette="Blues", hue="gender")
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

```python
#Estmos creando boxplots para comparar los puntajes de lectura y escritura por género, continuando el análisis que comenzamos con matemáticas.
#Esto nos permitirá hacer una comparación más completa del rendimiento académico.

plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='reading_score', data=df, palette="Greens", hue="gender")
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='writing_score', data=df, palette="Oranges", hue="gender")
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')
plt.show()
```

```python
#Estaremos calculando el promedio de cada columna
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

#Mostramos promedio por categoría
plt.figure(figsize=(12, 6))
sns.barplot(x='parental_level_of_education', y='average_score', data=df, palette="coolwarm")
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()
```

```python
#Podemos obeservar la:
#Comparación de distribución de puntajes
#Mediana (línea horizontal en la caja)
#Rango intercuartílico (tamaño de la caja)
#Valores atípicos (puntos fuera de los bigotes)

plt.figure(figsize=(10, 6))
sns.boxplot(x='lunch', y='average_score', data=df, palette="Set3", hue="lunch")
plt.title('Average Scores by Lunch Type')
plt.xlabel('Lunch Type (1 = Standard, 0 = Free/Reduced)')
plt.ylabel('Average Score')
plt.show()
```

```python
#Aspectos a observar:
#Comparación de puntajes entre estudiantes que tomaron/no tomaron curso
#Mediana (línea horizontal en la caja)
#Dispersión de los puntajes
#Posibles valores atípicos

plt.figure(figsize=(10, 6))
sns.boxplot(x='test_preparation_course', y='average_score', data=df, palette="Set2", hue="test_preparation_course")
plt.title('Average Scores by Test Preparation Course Completion')
plt.xlabel('Test Preparation Course (1 = Completed, 0 = Not Completed)')
plt.ylabel('Average Score')
plt.show()
```

```python
#Matriz de Correlación
#Incluye variables como género, almuerzo, preparación, puntajes en materias, puntaje total y promedio
#Valores van de -1 a 1
#1: Correlación positiva perfecta
#0: Sin correlación
#-1: Correlación negativa perfecta
correlation_matrix = df[['gender', 'lunch', 'test_preparation_course', 'math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].corr()
print(correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', vmin=0, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix of Student Performance Metrics')
plt.show()
```

```python
#Estamo realizando una prueba t de Student para comparar los puntajes de matemáticas entre géneros
#Prueba t independiente
#Compara medias de puntajes de matemáticas entre hombres y mujeres
#Usa la biblioteca SciPy para cálculos estadísticos

from scipy import stats
male_math_scores = df[df['gender'] == 1]['math_score']
female_math_scores = df[df['gender'] == 0]['math_score']
t_statistic, p_value = stats.ttest_ind(male_math_scores, female_math_scores) #Medir la diferencia entre ambos grupos
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

```python
#Comparamos puntajes promedio en 6 niveles de educación parental
#Usamos stats.f_oneway() para la prueba
#Determinamos si hay diferencias significativas entre los grupos
#F-statistic: Mide la varianza entre grupos vs dentro de grupos

anova_result = stats.f_oneway(
    df[df['parental_level_of_education'] == "high school"]['average_score'],
    df[df['parental_level_of_education'] == "associate's degree"]['average_score'],
    df[df['parental_level_of_education'] == "bachelor's degree"]['average_score'],
    df[df['parental_level_of_education'] == "master's degree"]['average_score'],
    df[df['parental_level_of_education'] == "some college"]['average_score'],
    df[df['parental_level_of_education'] == "some high school"]['average_score']
)
print(f"ANOVA result: F-statistic = {anova_result.statistic}, P-value = {anova_result.pvalue}")
```

### sns.boxplot()

##### Concepto básico:
-Representa distribución de datos numéricos
-Muestra resumen estadístico en una caja

1. Caja principal:
- Representa rango intercuartílico (25% - 75%)
- La línea horizontal dentro de la caja es la mediana

2. "Bigotes" (líneas verticales):
- Se extienden hasta 1.5 veces el rango intercuartílico
- Indican variabilidad de los datos

3. Puntos fuera de los bigotes:
- Considerados valores atípicos (outliers)
- Representan datos muy diferentes al resto

##### Parámetros principales:
- x: Variable categórica
- y: Variable numérica
- data: DataFrame
- hue: Permite subdividir por otra categoría
- palette: Esquema de colores

### sns.heatmap()

##### Concepto básico:
- Representa datos matriciales con color
- Útil para visualizar correlaciones, matrices de confusión

1. Matriz de datos
2. Escala de colores
3. Valores numéricos (opcional)

##### Parámetros clave:
- Matriz de datos
- annot=True: Muestra valores
- cmap: Esquema de colores
- vmin/vmax: Rango de valores
- linewidths: Separación entre celdas
